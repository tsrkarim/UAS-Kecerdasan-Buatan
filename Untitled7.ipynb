```python

import random
import matplotlib.pyplot as plt

# ==========================================
# 1. DATA INPUT SESUAI SOAL
# ==========================================
courses = [
    ("AI", "DosenA"), ("ML", "DosenB"), ("DB", "DosenC"),
    ("CN", "DosenD"), ("SE", "DosenE"), ("OS", "DosenF"),
    ("IR", "DosenA"), ("NLP", "DosenB"), ("CV", "DosenC"),
    ("IOT", "DosenD"), ("HCI", "DosenE"), ("CyberSec", "DosenF"),
    ("BigData", "DosenA"), ("DL", "DosenB"), ("NoSQL", "DosenC"),
    ("Cloud", "DosenD"), ("UX", "DosenE"), ("SecEng", "DosenF"),
    ("Agile", "DosenA"), ("RL", "DosenB"), ("ETL", "DosenC"),
    ("DevOps", "DosenD"), ("GameDev", "DosenE"), ("Blockchain", "DosenF")
]

rooms = ["R1", "R2", "R3", "R4"]

timeslots = [
    "Senin-08", "Senin-10", "Senin-13",
    "Selasa-08", "Selasa-10", "Selasa-13",
    "Rabu-08", "Rabu-10", "Rabu-13",
    "Kamis-08", "Kamis-10", "Kamis-13"
]

# Parameter Algoritma Genetika dari Soal
POP_SIZE = 60
GENERATIONS = 100
CROSSOVER_RATE = 0.85
MUTATION_RATE = 0.20

# ==========================================
# 2. EVALUASI FITNESS & KONFLIK
# ==========================================
def count_conflicts(individual):
    conflicts = 0

    # 1. Konflik Ruang: Dua MK memakai ruang & timeslot yang sama
    room_time_slots = {}

    # 2. Konflik Dosen-Waktu: Dosen mengajar >1 MK di timeslot yang sama
    teacher_time_slots = {}

    # 3. Konflik Dosen-Hari: Dosen mengajar >1 kali di hari yang sama
    teacher_days = {}

    for idx, (room, timeslot) in enumerate(individual):
        course, teacher = courses[idx]
        day = timeslot.split("-")[0]  # Ambil nama hari (misal: "Senin")

        # Cek Konflik Ruang
        rt_key = (room, timeslot)
        if rt_key in room_time_slots:
            conflicts += 1
        else:
            room_time_slots[rt_key] = True

        # Cek Konflik Dosen-Waktu
        tt_key = (teacher, timeslot)
        if tt_key in teacher_time_slots:
            conflicts += 1
        else:
            teacher_time_slots[tt_key] = True

        # Cek Konflik Dosen-Hari
        td_key = (teacher, day)
        if td_key in teacher_days:
            conflicts += 1
        else:
            teacher_days[td_key] = True

    return conflicts

def calculate_fitness(individual):
    conflicts = count_conflicts(individual)
    fitness = 100.0 / (1.0 + conflicts)
    return fitness, conflicts

# ==========================================
# 3. FUNGSI ALGORITMA GENETIKA
# ==========================================
def create_individual():
    return [(random.choice(rooms), random.choice(timeslots)) for _ in range(len(courses))]

def create_population(pop_size):
    return [create_individual() for _ in range(pop_size)]

def selection(population, fitnesses):
    # Tournament Selection
    tournament = random.sample(list(zip(population, fitnesses)), 3)
    tournament.sort(key=lambda x: x[1], reverse=True)
    return tournament[0][0]

def crossover(parent1, parent2):
    if random.random() < CROSSOVER_RATE:
        point = random.randint(1, len(courses) - 1)
        child1 = parent1[:point] + parent2[point:]
        child2 = parent2[:point] + parent1[point:]
        return child1, child2
    return parent1.copy(), parent2.copy()

def mutate(individual):
    for i in range(len(individual)):
        if random.random() < MUTATION_RATE:
            individual[i] = (random.choice(rooms), random.choice(timeslots))
    return individual

# ==========================================
# 4. PROSES EVOLUSI (MAIN LOOP)
# ==========================================
population = create_population(POP_SIZE)

max_fitness_history = []
avg_fitness_history = []

best_global_ind = None
best_global_fit = -1
best_global_conflicts = 999

# Seed acak agar repeatable
random.seed(42)

for gen in range(1, GENERATIONS + 1):
    evaluated = [calculate_fitness(ind) for ind in population]
    fitnesses = [f for f, c in evaluated]
    conflicts_list = [c for f, c in evaluated]

    max_fit = max(fitnesses)
    avg_fit = sum(fitnesses) / len(fitnesses)

    max_fitness_history.append(max_fit)
    avg_fitness_history.append(avg_fit)

    # Catat jadwal terbaik
    best_gen_idx = fitnesses.index(max_fit)
    if max_fit > best_global_fit:
        best_global_fit = max_fit
        best_global_ind = population[best_gen_idx]
        best_global_conflicts = conflicts_list[best_gen_idx]

    # Cetak perkembangan per generasi (Persis Format Soal)
    print(f"Gen | {gen}: Max Fitness = {max_fit:.2f} | Avg Fitness = {avg_fit:.2f}")

    # Buat Generasi Baru
    new_population = []
    # Elitisme: Simpan 2 individu terbaik
    sorted_pop = [x for _, x in sorted(zip(fitnesses, population), reverse=True)]
    new_population.extend(sorted_pop[:2])

    while len(new_population) < POP_SIZE:
        p1 = selection(population, fitnesses)
        p2 = selection(population, fitnesses)
        c1, c2 = crossover(p1, p2)
        new_population.append(mutate(c1))
        if len(new_population) < POP_SIZE:
            new_population.append(mutate(c2))

    population = new_population

# ==========================================
# 5. CETAK KELUARAN JADWAL TERBAIK
# ==========================================
print("\n=== JADWAL TERBAIK ===")
for idx, (room, timeslot) in enumerate(best_global_ind):
    mk, dosen = courses[idx]
    print(f"MK: {mk:10} | Ruang: {room} | Waktu: {timeslot:9} | Dosen: {dosen}")

print("\n=== HASIL EVALUASI AKHIR ===")
print(f"Fitness Terbaik: {best_global_fit:.2f}")
print(f"Jumlah Konflik : {best_global_conflicts}")

# ==========================================
# 6. VISUALISASI PROSES EVOLUSI
# ==========================================
plt.figure(figsize=(10, 6))
plt.plot(range(1, GENERATIONS + 1), max_fitness_history, label="Fitness Maksimum", color="black", linewidth=2)
plt.plot(range(1, GENERATIONS + 1), avg_fitness_history, label="Fitness Rata-rata", color="orange", linewidth=1.5)
plt.title("Perkembangan Fitness - Algoritma Genetika Penjadwalan Mata Kuliah")
plt.xlabel("Generasi")
plt.ylabel("Fitness")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
