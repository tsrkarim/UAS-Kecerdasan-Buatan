```python

import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# 1. PARAMETER BERDASARKAN NIM (24146047)
# ==========================================
NIM = "24146047"
nim_int = int(NIM)

# Parameter Algoritma Genetika
pop_size = 20           # Jumlah populasi (N)
chromosome_len = 10     # Panjang kromosom biner
generations = 50        # Jumlah generasi
pc = 0.8                # Crossover rate (80%)
pm = 0.05               # Mutation rate (5%)

# Batas nilai x untuk fungsi target (misal: -10 hingga 10)
x_min, x_max = -10, 10

# Seed acak menggunakan NIM agar hasil reproducible
np.random.seed(nim_int)

# ==========================================
# 2. FUNGSI UTAMA ALGORITMA GENETIKA
# ==========================================

# Fungsi Objective / Target yang ingin dimaksimalkan: f(x) = -x^2 + 4x + 10
def objective_function(x):
    return -(x**2) + (4 * x) + 10

# Decode Rantai Biner ke Nilai Real x
def decode_binary(chromosome, x_min, x_max):
    binary_str = "".join(map(str, chromosome))
    decimal_val = int(binary_str, 2)
    max_decimal = (2**len(chromosome)) - 1
    # Interpolasi linier ke rentang [x_min, x_max]
    x = x_min + (decimal_val / max_decimal) * (x_max - x_min)
    return x

# Inisialisasi Populasi Awal
def initialize_population(pop_size, chrom_len):
    return np.random.randint(0, 2, size=(pop_size, chrom_len))

# Seleksi Roda Roulette (Roulette Wheel Selection)
def roulette_wheel_selection(pop, fitness_values):
    # Penyesuaian agar fitness tidak negatif
    min_fit = min(fitness_values)
    adj_fitness = [f - min_fit + 1e-6 for f in fitness_values]
    total_fit = sum(adj_fitness)
    probs = [f / total_fit for f in adj_fitness]

    selected_indices = np.random.choice(len(pop), size=len(pop), p=probs)
    return pop[selected_indices]

# Pindah Silang (One-Point Crossover)
def crossover(parent1, parent2, pc):
    if np.random.rand() < pc:
        point = np.random.randint(1, len(parent1))
        child1 = np.concatenate([parent1[:point], parent2[point:]])
        child2 = np.concatenate([parent2[:point], parent1[point:]])
        return child1, child2
    return parent1.copy(), parent2.copy()

# Mutasi Bit (Bit-Flip Mutation)
def mutate(chromosome, pm):
    for i in range(len(chromosome)):
        if np.random.rand() < pm:
            chromosome[i] = 1 - chromosome[i]  # Flip 0 ke 1 atau 1 ke 0
    return chromosome

# ==========================================
# 3. PROSES ITERASI GENERASI (MAIN LOOP)
# ==========================================
population = initialize_population(pop_size, chromosome_len)

best_fitness_history = []
best_x_history = []

for gen in range(generations):
    # 1. Dekode dan Hitung Fitness
    x_values = [decode_binary(ind, x_min, x_max) for ind in population]
    fitness_values = [objective_function(x) for x in x_values]

    # Catat Individu Terbaik di Generasi Ini
    best_idx = np.argmax(fitness_values)
    best_fitness_history.append(fitness_values[best_idx])
    best_x_history.append(x_values[best_idx])

    # 2. Seleksi Induk
    selected_pop = roulette_wheel_selection(population, fitness_values)

    # 3. Reproduksi (Crossover & Mutasi)
    next_population = []
    for i in range(0, pop_size, 2):
        p1, p2 = selected_pop[i], selected_pop[i+1]
        c1, c2 = crossover(p1, p2, pc)
        c1 = mutate(c1, pm)
        c2 = mutate(c2, pm)
        next_population.extend([c1, c2])

    population = np.array(next_population)

# ==========================================
# 4. OUTPUT HASIL & GRAFIK KONVERGENSI
# ==========================================
global_best_idx = np.argmax(best_fitness_history)
print(f"=== HASIL OPTIMASI ALGORITMA GENETIKA (NIM: {NIM}) ===")
print(f"Solusi x Terbaik         : {best_x_history[global_best_idx]:.4f}")
print(f"Nilai Fitness Maksimum   : {best_fitness_history[global_best_idx]:.4f}")
print(f"Ditemukan pada Generasi  : {global_best_idx + 1}")

# Plot Grafik Konvergensi Fitness
plt.figure(figsize=(10, 5))
plt.plot(range(1, generations + 1), best_fitness_history, marker='o', color='b', linewidth=2)
plt.title(f"Grafik Konvergensi Fitness Algoritma Genetika (NIM: {NIM})")
plt.xlabel("Generasi")
plt.ylabel("Fitness Terbaik (Best Fitness)")
plt.grid(True)
plt.show()
```
